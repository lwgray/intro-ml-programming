### Idea 1: Customer support assistant

### Idea 2: News article processor
- Input: news article URL or text
- Pipeline: zero-shot classification (topic) → summarization (TL;DR) → LLM (suggest follow-up reading)
- Output: tagged article with summary and recommendations

### Idea 3: Image-based catalog generator
- Input: product photo
- Pipeline: CNN classifier or CLIP zero-shot (identify category) → LLM (generate description) → Stable Diffusion (generate styled variants for marketing)
