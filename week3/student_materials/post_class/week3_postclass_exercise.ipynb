## Part 8: Bonus Visualization (Optional)

**BONUS:** Create a 2D visualization of decision boundaries.

```python
# BONUS: Visualize decision boundaries using two most important features
# This is complex - code provided for learning!

# Use only two features for 2D visualization
# Based on feature importance, we'll use petal length and petal width
X_2d = X[:, [2, 3]]  # Indices 2 and 3 are petal length and petal width

# Split and scale
X_train_2d, X_test_2d, y_train_2d, y_test_2d = train_test_split(
    X_2d, y, test_size=0.2, random_state=42, stratify=y
)

# Train simple decision tree on 2D data
dt_2d = DecisionTreeClassifier(max_depth=5, random_state=42)
dt_2d.fit(X_train_2d, y_train_2d)

# Create mesh grid
h = 0.02
x_min, x_max = X_2d[:, 0].min() - 0.5, X_2d[:, 0].max() + 0.5
y_min, y_max = X_2d[:, 1].min() - 0.5, X_2d[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict on mesh
Z = dt_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, cmap='viridis', 
                     edgecolors='k', s=100)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Decision Tree Decision Boundaries (2D)')
plt.colorbar(scatter, ticks=[0, 1, 2], label='Species')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("The colored regions show what the Decision Tree predicts for each area.")
print("Notice how the boundaries are perpendicular lines (tree splits on axis-aligned thresholds).")
```
