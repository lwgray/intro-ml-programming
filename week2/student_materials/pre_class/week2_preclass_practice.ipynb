## Step 5: Make Predictions

**Two types of predictions in classification:**
1. **Hard predictions:** Class labels (0 or 1)
2. **Probability predictions:** Probability of each class (NEW!)

Let's see both:

````python
# TODO: Get class predictions (0 or 1)
# Hint: Use the predict() method on X_test
# Expected: Array of 0s and 1s

y_pred = model.____(____)  

# Probability predictions (provided)
y_pred_proba = model.predict_proba(X_test)

print("✅ Predictions complete!")
print(f"\nFirst 5 class predictions: {y_pred[:5]}")
print(f"\nFirst 5 probability predictions:")
print(y_pred_proba[:5])
````

**Understanding the output:**
- `y_pred`: Hard class labels (0 or 1)
- `y_pred_proba`: Two probabilities per sample:
  - Column 0: Probability of class 0
  - Column 1: Probability of class 1
  - They always sum to 1.0!

**Example:** `[0.1, 0.9]` means:
- 10% chance of class 0
- 90% chance of class 1
- So we predict class 1

````python
# Generate confusion matrix (provided)
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print("\nInterpretation:")
print(f"Top-left: Correctly predicted class 0")
print(f"Bottom-right: Correctly predicted class 1")
print(f"Top-right: Predicted 1, actually 0 (False Positive)")
print(f"Bottom-left: Predicted 0, actually 1 (False Negative)")
````

**Understanding the confusion matrix:**

```
                Predicted
                  0     1
              ┌─────┬─────┐
Actual     0  │ TN  │ FP  │
              ├─────┼─────┤
           1  │ FN  │ TP  │
              └─────┴─────┘
```

- **Diagonal (top-left to bottom-right):** Correct predictions!
- **Off-diagonal:** Errors

## Step 7: Visualize Confusion Matrix (Bonus)

**All code provided** - just run and observe!

````python
# Visualize confusion matrix
import seaborn as sns

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Class 0', 'Class 1'],
            yticklabels=['Class 0', 'Class 1'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
````

**Visual interpretation:**
- Darker blue = more samples
- Diagonal should be darkest (correct predictions)
- Off-diagonal should be light or white (few errors)
