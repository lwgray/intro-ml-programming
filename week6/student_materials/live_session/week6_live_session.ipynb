## Part 10: What Did the Network Actually Learn? 🔍

We told the "edges → shapes → digits" story as **intuition**. That story is literally true for **CNNs** (Week 8). But this network is **flat** — we threw away the 2D grid the moment we flattened 784 pixels.

Each neuron in the **first** hidden layer has 784 weights, one per pixel — and 784 = 28×28. So we can reshape any neuron's weights back into an image and *see* what it looks for.

> ⚠️ Only **layer 1** has a pixel picture. Layer 2's weights connect to the 128 hidden units (not pixels), so there's no 28×28 image of them. That missing picture *is* the point: the clean per-layer hierarchy is a CNN story, not a flat-MLP one.

```python
# What did the FIRST hidden layer actually learn?
# Each neuron in the first Dense layer has 784 weights - one per pixel.
# 784 weights = a 28x28 grid, so we can reshape and SEE what each neuron looks for.
W = model.layers[1].get_weights()[0]          # first Dense layer: shape (784, 128)

fig, axes = plt.subplots(2, 8, figsize=(12, 3.2))
for i, ax in enumerate(axes.ravel()):          # show 16 of the 128 neurons
    w = W[:, i].reshape(28, 28)
    lim = np.abs(w).max()
    ax.imshow(w, cmap='RdBu', vmin=-lim, vmax=lim)   # blue = +weight, red = -weight
    ax.axis('off')
fig.suptitle("What each layer-1 neuron 'looks for' (first-layer weights as 28x28)")
plt.tight_layout()
plt.show()
```

**What you're seeing:**
- Blobby, smudgy **stroke templates** — *not* clean edge detectors
- Each one is locked to a **specific location** in the frame (blue = "ink here", red = "no ink here")
- It works on MNIST only because the digits are centered and size-normalized

**The payoff:** Because we flattened the image, each neuron sees all 784 pixels with no idea which are neighbors — so it learns *edges-at-a-fixed-spot*, not reusable edges. Shift every digit a few pixels and this network struggles. **That is exactly the problem CNNs fix in Week 8.**
