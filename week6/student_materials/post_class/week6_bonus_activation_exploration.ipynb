## 🎓 Key Insights

**General findings from this type of experiment:**

- **ReLU:** Usually best for hidden layers (fast, no vanishing gradient)
- **tanh:** Can work but slower than ReLU, risk of vanishing gradient
- **sigmoid:** Often performs poorly in hidden layers (vanishing gradient)
- **LeakyReLU:** Similar to ReLU, slight advantage in some cases

**Why ReLU is the default:**
- Simple computation (max(0, x))
- No vanishing gradient problem
- Works well empirically

**When to try alternatives:**
- tanh: When you need outputs centered at 0
- LeakyReLU: When ReLU neurons "die" (output always 0)
- PReLU/ELU: Advanced alternatives (Week 7-8)
