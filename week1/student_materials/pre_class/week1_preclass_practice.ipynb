## Solution 5: Array Arithmetic

````python
a = np.array([10, 20, 30, 40, 50])
b = np.array([1, 2, 3, 4, 5])

sum_ab = a + b
diff_ab = a - b
double_a = a * 2
b_squared = b ** 2  # or: np.power(b, 2)

print(f"a + b = {sum_ab}")
print(f"a - b = {diff_ab}")
print(f"2 * a = {double_a}")
print(f"b² = {b_squared}")
````

---

## Solution 6: Indexing and Slicing

````python
arr = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])

first = arr[0]
last = arr[-1]
middle = arr[2:5]  # Indices 2, 3, 4 (end index is exclusive)
every_other = arr[::2]  # Start at 0, step by 2

print(f"First: {first}")
print(f"Last: {last}")
print(f"Middle: {middle}")
print(f"Every other: {every_other}")
````

## Solution 7: Array Statistics

````python
data = np.array([23, 45, 67, 12, 89, 34, 56, 78, 90, 11])

mean_val = np.mean(data)  # or: data.mean()
min_val = np.min(data)    # or: data.min()
max_val = np.max(data)    # or: data.max()
std_val = np.std(data)    # or: data.std()

print(f"Mean: {mean_val:.2f}")
print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Std Dev: {std_val:.2f}")
````

## Solution 8: Random Arrays

````python
np.random.seed(42)

random_ints = np.random.randint(1, 101, size=10)  # 1 to 100 inclusive
random_floats = np.random.randn(5)  # Standard normal distribution

print(f"Random integers: {random_ints}")
print(f"Random floats: {random_floats}")
````

---

## Solution 9: Boolean Indexing

````python
scores = np.array([45, 67, 89, 23, 78, 92, 56, 34, 81])

passing = scores[scores >= 70]
num_passing = (scores >= 70).sum()  # Boolean True = 1, False = 0

print(f"All scores: {scores}")
print(f"Passing scores (>=70): {passing}")
print(f"Number passing: {num_passing}")
````

---

## Solution 10: Calculating Errors

````python
y_actual = np.array([100, 150, 200, 250, 300])
y_pred = np.array([110, 140, 210, 240, 290])

errors = y_actual - y_pred
abs_errors = np.abs(errors)
squared_errors = errors ** 2
mse = np.mean(squared_errors)
rmse = np.sqrt(mse)

print(f"Actual:    {y_actual}")
print(f"Predicted: {y_pred}")
print(f"Errors:    {errors}")
print(f"Absolute errors: {abs_errors}")
print(f"Squared errors:  {squared_errors}")
print(f"\nMSE:  {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
````
