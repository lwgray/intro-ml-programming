## Part 1: Load & Prepare Data

```python
# Imports (provided)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
# Load wine dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

df = pd.DataFrame(X, columns=feature_names)
print(f"Dataset: {X.shape[0]} wines, {X.shape[1]} features")
df.head()
```

### TODO 1: Scale the features

**Why?** K-Means uses distances, so different scales would bias clustering.

**YOUR TASK:** Complete the code below.

```python
# SOLUTION: Create a StandardScaler and fit_transform X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("✅ Features scaled")
```

---

## Part 2: Choose K Using Multiple Methods

We'll use BOTH elbow method and silhouette scores.

### TODO 2: Complete the elbow method loop

```python
# Elbow method
inertias = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    # SOLUTION: Append kmeans.inertia_ to the inertias list
    inertias.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid(True, alpha=0.3)
plt.show()
```

**PAUSE:** Instructor will ask "Where do you see the elbow?"

### TODO 3: Calculate silhouette scores

```python
# Silhouette analysis
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    # SOLUTION: Calculate silhouette_score(X_scaled, labels)
    score = silhouette_score(X_scaled, labels)
    
    print(f"K={k}: Silhouette = {score:.3f}")
```

**PAUSE:** Instructor will discuss: "Which K has the highest silhouette? What does the score mean?"

---

## Part 3: Fit K-Means with Chosen K

Based on elbow + silhouette, we'll use **K=3**.

```python
# Fit K-Means with K=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Add to dataframe
df['cluster'] = clusters

print(f"Cluster distribution:\n{df['cluster'].value_counts().sort_index()}")
```

---

## Part 4: Interpret Clusters

What defines each cluster?

### TODO 4: Calculate cluster means

```python
# SOLUTION: Group by cluster and calculate mean for each feature
cluster_means = df.groupby('cluster').mean()

cluster_means
```

```python
# Visualize cluster characteristics
plt.figure(figsize=(12, 4))
sns.heatmap(cluster_means.T, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Cluster Characteristics')
plt.xlabel('Cluster')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

**PAUSE:** Instructor will ask: "What stands out? What defines Cluster 0 vs Cluster 1?"

---

## Part 5: PCA Visualization

We have 13 dimensions. Let's reduce to 2 for visualization.

### TODO 5: Apply PCA

```python
# SOLUTION: Create PCA with n_components=2 and fit_transform X_scaled
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(f"Variance explained by PC1: {pca.explained_variance_ratio_[0]:.1%}")
print(f"Variance explained by PC2: {pca.explained_variance_ratio_[1]:.1%}")
print(f"Total variance explained: {pca.explained_variance_ratio_.sum():.1%}")
```

```python
# Visualize clusters in PCA space
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.7, edgecolors='k')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
plt.title('Wine Clusters Visualized (PCA 2D)')
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**PAUSE:** Instructor will ask: "Are clusters well-separated? Do PCA boundaries match K-Means assignments?"

## Part 6: Reflection

**Connecting K-Means and PCA:**

### SOLUTION: Reflection Questions

**1. What's the difference between K-Means and PCA?**
- **K-Means:** A clustering algorithm that groups data points into K clusters based on similarity (minimizes within-cluster distances). It's about **finding groups**.
- **PCA:** A dimensionality reduction technique that transforms data to a new coordinate system where axes capture maximum variance. It's about **finding directions of variation**.

**2. Can you use PCA BEFORE K-Means? Why would you?**
- **Yes!** You can apply PCA first to reduce dimensions, then cluster in the lower-dimensional space. Reasons:
  - Reduce computational cost (clustering in 2D is faster than 13D)
  - Remove noise (minor variance components might be noise)
  - Avoid curse of dimensionality (distances become less meaningful in high dimensions)
  
**3. What % variance is "enough" for PCA?**
- **For visualization:** 50-80% is often acceptable. You're trying to see patterns, not preserve all information.
- **For preprocessing:** 90-95%+ is better. You want to retain most information before feeding to another algorithm.

**4. Is 55% variance explained "good" for this wine visualization?**
- **Answer:** It's acceptable but not great. We're losing 45% of information, so the 2D plot shows patterns but doesn't capture the full story. Some cluster separation might be hidden in the other 11 principal components. For exploratory visualization, this is OK - but we should acknowledge the limitation.
