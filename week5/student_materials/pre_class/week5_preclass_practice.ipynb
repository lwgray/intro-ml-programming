## Part 8: Key Takeaways

Before the live session, make sure you understand:

### ✅ Unsupervised Learning Is Different
- **No labels (y)** - you're discovering patterns, not predicting targets
- **Subjective interpretation** - "What does Cluster 2 mean?"
- **No single 'correct' answer** - different K values tell different stories

### ✅ K-Means Finds Natural Groupings
- Groups similar points together
- Each cluster has a center (centroid)
- Works well when clusters are spherical and well-separated

### ✅ Choosing K Is Important
- **Visual inspection:** Look at the data (when possible)
- **Elbow method:** Plot inertia vs K, find the bend
- **Domain knowledge:** Sometimes business needs dictate K

### ✅ sklearn Pattern Still Works!
```python
# Create model
model = KMeans(n_clusters=3)

# Fit (NO y!)
model.fit(X)

# Predict/assign clusters
clusters = model.predict(X)
```
